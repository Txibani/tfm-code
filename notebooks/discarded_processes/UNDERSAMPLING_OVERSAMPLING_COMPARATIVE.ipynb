```python
# Librerías de manipulación de datos
import pandas as pd
import numpy as np

# Librerías de visualización de datos
import matplotlib.pyplot as plt
import seaborn as sns

# Librerías de procesado de datos
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve, auc
import imblearn
from imblearn.over_sampling import SMOTE, SMOTENC
from imblearn.under_sampling import NearMiss
from sklearn.preprocessing import MinMaxScaler
from imblearn.pipeline import Pipeline
from imblearn.ensemble import BalancedBaggingClassifier

# Otras librerías
import warnings
warnings.filterwarnings('ignore')
```

# 1. Carga de los datos

```python
# Carga de los datos
file_path = "../data/DataBase.dat"
data = pd.read_csv(file_path, delim_whitespace = True)
data.head(5)
```

# 2. Procesado de los datos

Se están borrando estas variables debido a que la correlación entre estas y la variable Episode_ID es muy alta (comprobar la correlación en el notebook ORIGINAL_DATASET_ANALYSIS.ipynb)

```python
df = data.drop(['ID', 'Time1', 'Time2', 'Timetotal'], axis = 1)
```

```python
# Refactorizamos las variables al tipo adecuado
df['Episode_ID'] = df['Episode_ID'].astype('float')
df['Age'] = df['Age'].astype(int)
df['Family_History'] = df['Family_History'].astype('float')
df['BBD_History'] = df['BBD_History'].astype('float')
df['Mammographic_Features'] = df['Mammographic_Features'].astype('float')
df['BreastCancer'] = df['BreastCancer'].astype('float')
```

```python
# Se aplica la codificación one-hot encoder a todas aquellas variables categóricas con más
# de dos clases para convertirlas a tipo numérico
onehot_encoder = OneHotEncoder()

episode_id_df = pd.DataFrame(onehot_encoder.fit_transform(df[['Episode_ID']]).toarray())
episode_id_df.columns = ['episode_1', 'episode_2', 'episode_3', 'episode_4',
                          'episode_5', 'episode_6', 'episode_7', 'episode_8', 'episode_9', 'episode_10']

bbd_history_df = pd.DataFrame(onehot_encoder.fit_transform(df[['BBD_History']]).toarray())
bbd_history_df.columns = ['bbd_none', 'bbd_unknown', 'bbd_no_proliferative', 'bbd_proliferative']

mammographic_features_df = pd.DataFrame(onehot_encoder.fit_transform(df[['Mammographic_Features']]).toarray())
mammographic_features_df.columns = ['mf_none', 'mf_mass', 'mf_calcification', 'mf_asymetric_density',
                           'mf_distorsion', 'mf_multiple_patron']

df = df.drop(['Episode_ID', 'BBD_History', 'Mammographic_Features'], axis = 1)

df = pd.concat([df, episode_id_df, bbd_history_df, mammographic_features_df], axis = 1)

df.head(10)
```

```python
# Se escala la variable Age
final_df = df
scaler = MinMaxScaler()
final_df['Age'] = scaler.fit_transform(final_df[['Age']])
final_df.head()
```

# 3. Técnicas de procesado del dataset


- 1. Sin proceso
- 2. Parámetro class_weight = 'balanced'
- 3. Undersampling
- 4. Oversampling (SMOTE)
- 5. Mezcla de undersampling y oversampling

(Nota: solo resultados para el árbol de decisión)

```python
# Repartición de los datos
X = final_df.iloc[:, final_df.columns != 'BreastCancer'].values.astype(float)
y = final_df.iloc[:, final_df.columns == 'BreastCancer'].values.astype(float)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25, random_state=21)
```

### 1. Sin proceso

```python
# Definición del modelo
model = DecisionTreeClassifier(random_state=12)

# Entrenamiento del modelo
model.fit(X_train, y_train)

# Predicción sobre test
y_pred_test = model.predict(X_test)

# Evaluación sobre test
print(confusion_matrix(y_test, y_pred_test))
sns.heatmap(confusion_matrix(y_test, y_pred_test), annot = True)

print(classification_report(y_test, y_pred_test))
```

### 2. Parámetro class_weight = 'balanced'

```python
# Definición del modelo
model = DecisionTreeClassifier(random_state=12, class_weight='balanced')

# Se entrena el modelo
model.fit(X_train, y_train)

# Predicción sobre test
y_pred_test = model.predict(X_test)

# Evaluación sobre test
print(confusion_matrix(y_test, y_pred_test))
sns.heatmap(confusion_matrix(y_test, y_pred_test), annot = True)

print(classification_report(y_test, y_pred_test))
```

### 3. Undersampling

```python
# Técnica de undersampling NearMiss
nr = NearMiss()
X_train_nr, y_train_nr = nr.fit_resample(X_train, y_train)

model = DecisionTreeClassifier(random_state=12)
model.fit(X_train_nr, y_train_nr)

# Test sobre los datos originales
y_pred_test = model.predict(X_test)

print(confusion_matrix(y_test, y_pred_test))
sns.heatmap(confusion_matrix(y_test, y_pred_test), annot = True)

print(classification_report(y_test, y_pred_test))
```

### 4. Oversampling

```python
# SMOTE sobre los datos de entrenamiento
smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

model = DecisionTreeClassifier(random_state=12)
model.fit(X_train_smote, y_train_smote)

# Test sobre los datos originales
y_pred_test = model.predict(X_test)

print(confusion_matrix(y_test, y_pred_test))
sns.heatmap(confusion_matrix(y_test, y_pred_test), annot = True)

print(classification_report(y_test, y_pred_test))
```

#### 5. Mezcla de undersampling y oversampling

```python
# Primero se aplica oversampling
over = SMOTE(sampling_strategy=0.1)

# Posteriormente se aplica  undersampling
under = NearMiss(sampling_strategy=0.5)

# Pipeline para concatenar ejecuciones
steps = [('o', over), ('u', under), ('m', DecisionTreeClassifier(random_state=12))]
pipeline = Pipeline(steps=steps)

# Se enterna el modelo
pipeline.fit(X, y)

# Test sobre los datos originales
y_pred_test = pipeline.predict(X_test)

print(confusion_matrix(y_test, y_pred_test))
sns.heatmap(confusion_matrix(y_test, y_pred_test), annot = True)

print(classification_report(y_test, y_pred_test))
```

## Resultado final

El mejor resultado es al aplicar SMOTE ya que es el que mejor predice las muestras positivas y negativas.
